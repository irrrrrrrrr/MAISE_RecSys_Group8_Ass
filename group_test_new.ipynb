```python
import pandas as pd
import group_recommender_algs_new as gra

ratings_data = pd.read_csv("Dataset/last/XWines_Slim_150K_ratings.csv")
wine_data = pd.read_csv("Dataset/last/XWines_Slim_1K_wines.csv")
merged_data = pd.merge(ratings_data, wine_data, on='WineID')
wine_data = gra.avg_ratings(wine_data,ratings_data)
group_data = pd.read_csv("Dataset/last/group_composition.csv")
```

```python
group_id = 1
```

```python
group_info = group_data[group_data['group_id'] == group_id].iloc[0]
group_members = eval(group_info['group_members']) 
display(group_info)
```

```python
# ratings given by first two group members - just for example
group_ratings0 = merged_data[merged_data['UserID'] == group_members[0]]
display(group_ratings0.sort_values(by=['UserID', 'Rating'], ascending=[True, False]))
group_ratings1 = merged_data[merged_data['UserID'] == group_members[1]]
display(group_ratings1.sort_values(by=['UserID', 'Rating'], ascending=[True, False]))
```

```python
recommended_wines = gra.group_rec(group_id, group_data, merged_data,wine_data,ratings_data, None)
display(recommended_wines)
```

```python
from lenskit.algorithms import Recommender
from lenskit.algorithms.item_knn import ItemItem

itemitem = ItemItem(15, min_nbrs=3)  
recsys_item = Recommender.adapt(itemitem)
recsys_item.fit(merged_data.rename(columns={'UserID':'user','WineID':'item','Rating':'rating'}))
```

```python
recommended_wines_item = gra.group_rec(group_id, group_data, merged_data,wine_data,ratings_data, recsys_item)
display(recommended_wines_item)
```

```python
from lenskit.algorithms import Recommender
from lenskit.algorithms.user_knn import UserUser

useruser = UserUser(15, min_nbrs=3)  
recsys_user = Recommender.adapt(useruser)
recsys_user.fit(merged_data.rename(columns={'UserID':'user','WineID':'item','Rating':'rating'}))
```

```python
recommended_wines_user = gra.group_rec(group_id, group_data, merged_data,wine_data,ratings_data, recsys_user)
display(recommended_wines_user)
```
