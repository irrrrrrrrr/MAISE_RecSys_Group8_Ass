# Wine recommender 

## Load the data

```python
wines_df = pd.read_csv('Dataset/last/XWines_Slim_1K_wines.csv', low_memory=False, index_col='WineID')
ratings_df = pd.read_csv('Dataset/last/XWines_Slim_150K_ratings.csv', low_memory=False)
```

## Compute the average rating for each wine

```python
average_ratings = ratings_df.groupby('WineID')['Rating'].mean().reset_index()
wine_df_avg_ratings = pd.merge(wines_df, average_ratings, on='WineID', how='left')
wine_df_avg_ratings.rename(columns={'Rating': 'AvgRating'}, inplace=True)
wine_df_avg_ratings.set_index('WineID', inplace=True)
wine_df_avg_ratings
```

## Aggregation algorithms

```python
def get_wines_ratings(wine_id):
    ratings = ratings_df[ratings_df.WineID == wine_id].reset_index(drop=True).copy(deep=True)
    return ratings
```

```python
get_wines_ratings(103471)
```

```python
def get_wine_info(wine_id):
    info = wine_df_avg_ratings.loc[wine_id].copy(deep=True)
    return info
```

```python
def get_rated_wines(user_id):
    rated_wines = ratings_df[ratings_df.UserID == user_id].reset_index(drop=True).copy(deep=True)
    return rated_wines
```

### Create a dataframe with the scores of the wines. Initial values are 0

```python
wine_score = (pd.DataFrame({
        'WineID': wine_df_avg_ratings.index,
        'Score': 0 
}))
wine_score.set_index('WineID', inplace=True)
wine_score
```

### Establish the weights

```python
# Adjust the provided code and execute it with the necessary data
def recommend_wines(user_id, wine_ratings, wine_info, top_n=10, user_weight=0.7, avg_weight=0.3):
    """
    Recommends wines for a specific user based on their previous ratings.
    The recommendations are scaled between 0 and 10.
    
    Parameters:
    - user_id: the ID of the user for whom to make recommendations.
    - wine_ratings: DataFrame containing user ratings for different wines.
    - wine_info: DataFrame containing wine information.
    - top_n: Number of top recommendations to return.
    - user_weight: The weight given to the user's rating in the recommendation score.
    - avg_weight: The weight given to the average rating of a wine across all users.
    
    Returns:
    - A DataFrame with the top N recommended wines and their likelihood score.
    """
    # Step 1: Get the user's previous ratings
    user_ratings = wine_ratings[wine_ratings['UserID'] == user_id]
    
    if user_ratings.empty:
        return wine_info[['WineName', 'Type', 'Grapes', 'RegionName', 'AvgRating']].sort_values(by='AvgRating', ascending=False).head(top_n)
        
    # Step 2: Merge the user's ratings with the wine information
    wine_with_user_ratings = pd.merge(wine_info, user_ratings[['WineID', 'Rating']], on='WineID', how='left', suffixes=('', '_user'))
    
    # Step 3: Compute the likelihood score
    # Initialize a new column for LikelihoodScore
    wine_with_user_ratings['LikelihoodScore'] = 0
    
    # Loop over each row and calculate the likelihood score
    for idx, row in wine_with_user_ratings.iterrows():
        if pd.notna(row['Rating']):  # If the user has rated this wine
            # Calculate the weighted average of the user's rating and the average rating
            likelihood_score = (row['Rating'] * user_weight) + (row['AvgRating'] * avg_weight)
        else:
            # If the user hasn't rated this wine, use the average rating (scaled to 0-10)
            likelihood_score = (row['AvgRating'] * 10) / 5
    
        # Assign the calculated likelihood score to the corresponding row
        wine_with_user_ratings.at[idx, 'LikelihoodScore'] = likelihood_score
    
    # Step 4: Return the top N recommendations
    recommendations = wine_with_user_ratings[['WineName', 'Type', 'Grapes', 'RegionName', 'AvgRating', 'LikelihoodScore']].sort_values(by='LikelihoodScore', ascending=False).head(top_n)
    
    return recommendations

# Test the function for a specific user (using user ID 1209683 as an example)
user_id = 1209683
recommendations = recommend_wines(user_id, ratings_df, wine_df_avg_ratings)

recommendations
```
