```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

metrics_df = pd.read_csv("user_explanation_metrics.csv")
metrics_df.head()
```

```python
metrics_to_plot = ['Explanation Coherence', 'Diversity']

for metric in metrics_to_plot:
    plt.figure(figsize=(8, 4))
    sns.histplot(metrics_df[metric], bins=20, kde=True, color='skyblue')
    plt.title(f'Distribution of {metric} Across All Users')
    plt.xlabel(metric)
    plt.ylabel('Frequency')
    plt.grid(axis='y')
    plt.show()
```
